#### *Importing Libraries*

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

#### *Reading CSV to Pandas Dataframe*

```python
df=pd.read_csv("..\data\Carbon Emission.csv")
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
categorical_features = df.select_dtypes(include=['object']).columns
numerical_features = df.select_dtypes(include=['int64', 'float64']).columns

print("Categorical:", categorical_features)
print("Numerical:", numerical_features)
```

#### *Filling Missing Values*

```python
# Fill categorical missing values with the most frequent category
for col in categorical_features:
    df[col] = df[col].fillna(df[col].mode()[0])

# Fill numeric missing values with median
for col in numerical_features:
    df[col] = df[col].fillna(df[col].median())
```

```python
df.isnull().sum()
```

#### *Check For Duplicates*

```python
df.duplicated().sum()
```

```python
df.nunique()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['CarbonEmission'], bins=30, kde=True)
plt.title('Distribution of Carbon Emission')
plt.show()
```

#### *HeatMap*

```python
plt.figure(figsize=(10,6))
sns.heatmap(df[numerical_features].corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()

```

```python
df["Recycling"] = df["Recycling"].astype(str)
df["Cooking_With"] = df["Cooking_With"].astype(str)
```

#### *Saving The Cleaned File*

```python
df.to_csv("../data/Cleaned_Carbon_Emission.csv", index=False)
```
